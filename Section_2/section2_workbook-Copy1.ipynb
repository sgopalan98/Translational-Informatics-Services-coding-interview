```python
max_duplications = 0
for col in df:
    series = df[col]
    value_counts = series.value_counts()

    # Find the maximum count of duplications
    max_duplications = max(value_counts.max(), max_duplications)
max_duplications
```

```python
import pandas as pd
# example


def quant_norm(df_input):
    df = df_input.copy()
    ranks = (df.rank(method="first")
              .stack())
    rank_mean = (df.stack()
                   .groupby(ranks)
                   .mean())
    # Add interpolated values in between ranks
    finer_ranks = ((rank_mean.index + (1.0 / max_duplications)).to_list() +
                    rank_mean.index.to_list())
    rank_mean = rank_mean.reindex(finer_ranks).sort_index().interpolate()
    return (df.rank(method='average')
              .stack()
              .map(rank_mean)
              .unstack())


display(df)
quant_norm_df = quant_norm(df)
display(quant_norm_df)
```

```python
for col in quant_norm_df:
    print(quant_norm_df[col].sum())
```
